##Create the table

```python
%sql
CREATE TABLE IF NOT EXISTS cw_ucc_services_archer_amr_cop_mckesson_d.control.file_ingestion_config (
    id INT,

    -- SOURCE
    source_file_name STRING,
    source_path STRING,
    file_type STRING,                 -- csv, xlsx, xlsm
    sheet_name STRING,                -- NULL for CSV

    -- TARGET
    target_catalog STRING,
    target_schema STRING,
    target_table STRING,

    -- PIPELINE METADATA
    pipeline_name STRING,
    source_system STRING,

    -- INGESTION BEHAVIOR
    ingestion_strategy STRING,        -- overwrite, append
    delimiter STRING,                -- for CSV
    header BOOLEAN,
    infer_schema BOOLEAN,

    -- EXECUTION CONTROL (KEEP THIS)
    load_frequency STRING,            -- daily, weekly, monthly

    -- GENERAL CONTROL
    is_active BOOLEAN,
    load_order INT,

    -- AUDIT
    created_at TIMESTAMP,
    updated_at TIMESTAMP
);
```
