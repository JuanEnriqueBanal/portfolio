```python
%sql
CREATE TABLE IF NOT EXISTS cw_ucc_services_archer_amr_cop_mckesson_d.control.file_ingestion_log (
    run_id STRING,

    config_id INT,
    source_file_name STRING,
    sheet_name STRING,

    target_catalog STRING,
    target_schema STRING,
    target_table STRING,

    load_frequency STRING,

    status STRING,                 -- SUCCESS / FAILED
    row_count INT,

    start_time TIMESTAMP,
    end_time TIMESTAMP,

    error_message STRING,

    created_at TIMESTAMP
);
```

```python
from uuid import uuid4

RUN_ID = str(uuid4())
```

```python
%sql
SELECT *
FROM cw_ucc_services_archer_amr_cop_mckesson_d.control.file_ingestion_log
ORDER BY created_at DESC;
```

```python
%sql
SELECT
    target_catalog,
    target_schema,
    target_table,
    status,
    COUNT(*) AS execution_count,
    SUM(row_count) AS total_rows_loaded,
    MAX(end_time) AS last_execution_time
FROM cw_ucc_services_archer_amr_cop_mckesson_d.control.file_ingestion_log
GROUP BY
    target_catalog,
    target_schema,
    target_table,
    status
ORDER BY last_execution_time DESC;
```
